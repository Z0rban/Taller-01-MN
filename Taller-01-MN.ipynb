```python
def eqn(x:int) -> int:
    return (x**3)-(7*x**2)+4*x+12
```

```python
import numpy as np


##Genera 20 datos del -10 al 10
Xs, step = np.linspace(-10,10,num=20,retstep=True)
```

```python
Ys= []
for x in Xs:
    y = eqn(x)
    Ys.append(y)
    print(f"x= {x:.2f} \ty= {y:.2e}")
Ys
```

```python
from matplotlib import pyplot as plt

#import seabor as sns ## alternativa

_ = plt.plot(Xs,Ys, marker="o", color="red", linestyle=":")
_ = plt.xlabel("Eje x")
_ = plt.ylabel("Eje y")
plt.title("Gráfica de función cúbica")
plt.legend(["$y(x)=x^3-7x^2+4x+12$"])
plt.grid()
plt.show()

## TAREA HACER ZOOM
```
